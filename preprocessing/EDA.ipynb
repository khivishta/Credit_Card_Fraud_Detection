# open and visualise the csv file

```python
file_path = "../data/creditcard_2023.csv"
```

```python
# Read the CSV file into a pandas DataFrame
data = pd.read_csv(file_path)

# Display the column headers and the first few rows of the DataFrame
data.head()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# 1. Display Basic Information
print("Basic Information about the DataFrame:")
print(data.info())

# 2. Summary Statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# 3. Class Distribution
class_distribution = data["Class"].value_counts()
print("\nClass Distribution:")
print(class_distribution)
```

```python
# 4. Correlation Analysis (if features are anonymized)
correlation_matrix = data.corr()
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# 5. Visualization - Example: Histograms
data.hist(figsize=(22, 19), bins=20)
plt.suptitle("Histograms of Features")
plt.show()
```
